```python
import spacy
from spacy.matcher import Matcher
nlp = spacy.load("en_core_web_sm")
```

```python
class Equivalence:

    def __init__(self,path,save_path):
        with open(path) as f:
          text = f.readlines()
        self.en_li = []
        self.zh_li = []
        for i, sent in enumerate(text):
            if i%2 == 0:
                self.en_li.append(sent.strip())
            elif i%2 == 1:
                self.zh_li.append(sent.strip())
        self.save_path = save_path


    def NER(self):
        e = 1
        matcher = Matcher(nlp.vocab)
        pattern = [{"TEXT": {"REGEX": "^[A-Z]{2,}$"}}]
        matcher.add("AllCapsAbbreviation", [pattern])
        
        for en,zh in zip(self.en_li[400000:], self.zh_li[400000:]):
            print(e)
            e += 1
            doc = nlp(en)
            matches = matcher(doc)
            ner_li = []
            for match_id, start, end in matches:
                matched_text = doc[start:end].text
                
                if matched_text not in zh:
                    ner_li.append(matched_text+':')
            for pair in list(set(ner_li)):
                print(en)
                print(zh)
                print(pair)
                with open(save_path,'a+') as y:
                    y.write(en+'\n'+zh+'\n'+pair+'\n')



spath = '/content/gdrive/MyDrive/Colab Notebooks/literal translationese/dataset/Bilingual/Education/Bi-Education.txt'
save_path = '/content/gdrive/MyDrive/Colab Notebooks/literal translationese/dataset/feature1_equivalence/NER_2.txt'             
        
ptcl = Equivalence(spath, save_path)

ptcl.NER()
            
            
```
