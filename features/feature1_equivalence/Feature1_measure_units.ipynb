```python

zh_text_path = "/content/gdrive/MyDrive/Colab Notebooks/literal translationese/dataset/Chinese_text/Bi-Microblog_zh.txt"
en_text_path = "/content/gdrive/MyDrive/Colab Notebooks/literal translationese/dataset/English_text/Bi-Microblog_en.txt"
save_path = "/content/gdrive/MyDrive/Colab Notebooks/literal translationese/dataset/features_equivalence_NER/NER.txt"

with open(zh_text_path) as f:
    text_zh = f.readlines()
with open(en_text_path) as g:
    text_en = g.readlines()

for sent_zh,sent_en in zip(text_zh,text_en):
    pos = posseg.cut(sent_zh)
    for ph in pos:
        label = str(ph).split("/")[1]
        if label == "q" and str(ph).split("/")[0] in ['公里','里','公斤','斤','亩','公亩','尺','寸','斗','两','丈','平','担']:
            print(sent_en)
            print(sent_zh)
            print(ph)
            print("----------------------")

  

```
