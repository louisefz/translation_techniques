```python
import spacy
from idiomatch import Idiomatcher
nlp_en = spacy.load("en_core_web_sm")

from jieba import posseg
idiomatcher = Idiomatcher.from_pretrained(nlp_en)
```

```python
sent = "A blessing in disguise"  # a usecase of *open the floodgates*
  # this will take approx 50 seconds.
doc = nlp_en(sent)  # process the sentence with an nlp pipeline
print(idiomatcher.identify(doc))  # identify the idiom in the sentence

```

```python
class Generalization:

    def __init__(self,path,save_path):
        with open(path) as f:
          text = f.readlines()
        self.en_li = []
        self.zh_li = []
        for i, sent in enumerate(text):
            if i%2 == 0:
                self.en_li.append(sent.strip())
            elif i%2 == 1:
                self.zh_li.append(sent.strip())
        self.save_path = save_path

    def idiom(self):
        e = 1
        for en,zh in zip(self.en_li, self.zh_li):
            doc = nlp_en(en)
            en_idiom = idiomatcher.identify(doc)
            if en_idiom != []:
                print(en)
                print(zh)
                print(en_idiom)
                print('===================')


spath = '/content/gdrive/MyDrive/Colab Notebooks/literal translationese/dataset/Bilingual/Education/Bi-Education.txt'
save_path = '/content/gdrive/MyDrive/Colab Notebooks/literal translationese/dataset/feature45_particular_generalization/general_idiom.txt'             
        
ptcl = Generalization(spath, save_path)

ptcl.idiom()
```
